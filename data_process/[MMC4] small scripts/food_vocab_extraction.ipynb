```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import string
import csv


nltk.download('punkt')
```

```python
latin_prod = pd.read_csv('latin_products.tsv', sep="\t", skipinitialspace=True)
latin_prods = latin_prod["0"]
```

```python
latin_prods[0]
```

```python
string.punctuation
```

```python
latin_prods = latin_prods.to_numpy()
```

```python

stripPunct = str.maketrans('', '', string.punctuation)
vecTrans=np.vectorize(lambda x:x.translate(stripPunct))
latin_prods = vecTrans(latin_prods)
latin_prods
```

```python
latin_prods_tokenized = [word_tokenize(s.lower()) for s in latin_prods]
```

```python
latin_prods_tokenized
```

```python
stop_words = stopwords.words()
vocab = [w for s in latin_prods_tokenized for w in s if w not in stop_words]
vocabulary, counts = zip(*Counter(vocab).most_common())
```

```python
word_to_ind_full = {w: i for i,w in enumerate(vocabulary)}
ind_to_word_full = dict(enumerate(vocabulary))
ind_to_freq_full = dict(enumerate(counts))
```

```python
vocab_cnt = Counter(vocab).most_common()
print(len(vocab_cnt))
vocab_cnt = [c for c in vocab_cnt if c[1] >2]
print(len(vocab_cnt))
vocab_cnt
```

```python
with open('vocab_refined.csv', 'w', encoding="utf-8") as f:
    writer = csv.writer(f , lineterminator='\n')
    for tup in Counter(vocab).most_common():
        writer.writerow(tup)
```

```python
with open('vocab_refined_2.csv', 'w', encoding="utf-8") as f:
    writer = csv.writer(f , lineterminator='\n')
    for tup in vocab_cnt:
        writer.writerow(tup)
```

```python
vocab, counts = zip(*vocab_cnt)
vocab_list = list(vocab)
vocab_list
```

```python
with open(r'../data_preprocessing/vocab_word_list.txt', 'w', encoding="utf-8") as f:
    for i in vocab_list:  
        f.write("%s " % i)
```

```python
with open(r'../data_preprocessing/vocab_word_list_no_numbers.txt', 'w', encoding="utf-8") as f:
    for i in vocab_list:
        if not i.isnumeric():
            f.write("%s " % i)
```
