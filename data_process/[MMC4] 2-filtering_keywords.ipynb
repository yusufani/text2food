```python
with open("2_class_classification_results.json", "r", encoding="utf-8") as f:
    data = json.loads(f.read())
```

```python
data
```

```python
len(data)
```

```python
percentage = 90

listt= []
for i in data:
  if i['labels'][0] == 'food related' and i['scores'][0] >= percentage/100 and not any(x.isdigit() for x in i['sequence']):
        print(i['sequence'], i['scores'])
        listt.append(i['sequence'])

```

```python
listt
```

```python
len(listt)
```

```python
with open(f'more_than_{percentage}_classified_food_words.txt', 'w', encoding="utf-8") as f:
    for i in listt:
        f.write("%s " % i)
```

```python
# for the three classes
with open("3_class_classification_results.json", "r", encoding="utf-8") as f:
    data = json.loads(f.read())
```

```python
data
```

```python
percentage = 95

listt= []
for i in data:
  if any(x >= percentage/100 for x in i['scores']) and not any(x.isdigit() for x in i['sequence']):
        print(i['sequence'], i['scores'])
        listt.append(i['sequence'])
```

```python
len(listt)
```

```python
with open(f'3_classes_more_than_{percentage}_classified_food_words.txt', 'w', encoding="utf-8") as f:
    for i in listt:
        f.write("%s " % i)
```
