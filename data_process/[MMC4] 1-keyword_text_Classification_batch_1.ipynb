```python
from transformers import pipeline
import torch
from tqdm import tqdm
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=device)
```

```python
with open("word_list1.txt", 'r', encoding="utf-8") as file:
  word_list_string = file.read()
word_list = word_list_string.split()
```

```python
word_list
```

```python
len(word_list)
```

```python
candidate_labels = ['food related', 'non food related']
```

```python
results = []
bs = 1024
for i in tqdm(range(0, len(word_list), bs)):
    batch = [i for i in word_list[i: i + bs]]
    res = classifier(batch, candidate_labels)
    results.extend(res)
```
