```python
valids = []
not_valids = []

from tqdm import tqdm

def process_row(scores,labels):
  positives = 0.0
  negative = 0.0
  for i,label in enumerate(labels):
    if label == 'describing food' or label =='describing beverages' :
      positives += float(scores[i])
    elif label == 'not describing food' or label == 'not describing beverages':
      negative += float(scores[i])
    else:
      raise Exception(f'{label} not found' )

  return positives

for idx, row in tqdm(df.iterrows() , total=len(df)):

  positives = process_row(row['scores'] , row['labels'])
  if positives > 0.80:
    valids.append(row)
  else:
    not_valids.append(row)
    #if (image_folder / row['image_name'] ).exists():
      #print(image_folder / row['image_name'])
      #a =1
      #(image_folder / row['image_name'] ).unlink(missing_ok=True)

#%%
df_valid = pd.DataFrame(valids)
df_not_valid = pd.DataFrame(not_valids)
```
